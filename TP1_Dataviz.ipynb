### Indications
##### bornes :
plt.axis([xmin,xmax,ymin,ymax]) 
=> spécifier les bornes de la représentation, ce qui permet un zoom, éviter les grandes valeurs des fonctions..

##### couleur du trait :
pour changer la couleur du tracé une lettre g vert (green), r rouge (red), k noir, b bleu, c cyan, m magenta, y jaune (yellow), w blanc (white).

plt.plot(np.sin(x),'r') tracera la courbe sinus en rouge.
 
RQ: Pour avoir encore plus de couleurs, comme en HTML, utiliser la séquence color='#eeefff' ou aussi color=( R, G, B) avec trois paramètres compris entre 0 et 1 .

##### symboles :
mettre des symboles aux points tracés se fait via l’option marker. Les possibilités sont nombreuses parmi [ ‘+’ | ‘*’ | ‘,’ | ‘.’ | ‘1’ | ‘2’ | ‘3’ | ‘4’ | ‘<’ | ‘>’ | ‘D’ | ‘H’ | ‘^’ | ‘_’ | ‘d’ | ‘h’ | ‘o’ | ‘p’ | ‘s’ | ‘v’ | ‘x’ | ‘|’ | TICKUP | TICKDOWN | TICKLEFT | TICKRIGHT | ‘None’ | ‘ ‘ | ‘’ ].

##### style du trait :
pointillés, absences de trait, etc avec linestyle. Au choix '-' ligne continue, '--' tirets, '-.' points-tirets, ':' pointillés, sachant que 'None', '', ' ' donnent “rien-du-tout”. Plutôt que linestyle, ls (plus court) fait le même travail.

##### épaisseur du trait :
linewidth=flottant (comme linewidth=2) ,ou aussi lw, donne un trait, pointillé (tout ce qui est défini par style du trait) 

##### taille des symboles (markers) : 
markersize=flottant comme pour l’épaisseur du trait. 
D’autres paramètres sont modifiables markeredgecolor la couleur du trait du marker, markerfacecolor la couleur de l’intérieur (si le marker possède un intérieur comme 'o'), markeredgsize=flottant l’épaisseur du trait 

##### titre : 
plt.title("Mon titre")

* Tracer l'histogramme des probabilités suivant

```python
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

mu, sigma = 100, 15
x = mu + sigma*np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='green', alpha=0.75)

# add a 'best fit' line
y = mlab.normpdf( bins, mu, sigma)
l = plt.plot(bins, y, 'r--', linewidth=1)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title(r'$\mathrm{Histogram\ of\ IQ:}\ \mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)

plt.show()
```
